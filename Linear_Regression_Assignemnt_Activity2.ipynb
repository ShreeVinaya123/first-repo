Problem Statement: A real estate company wants to predict house prices based on the size of the house (in square feet). You are given the following data: House Size (sq ft) Price (in $1000) 600 150 1000 200 1200 240 1400 280 1800 310


Task 1: Fit a Linear Regression Model (5 Marks)
                

(Hints: Use a dataset that contains two numerical columns: one representing the independent variable (e.g., "House Size ") and the other representing the dependent variable (e.g., "Price ").

```python
#import modules
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Create house price dataset
house_data = {
    'House_Size': [600, 1000, 1200, 1400, 1800],
    'Price': [150, 200, 240, 280, 310]
}

df_house = pd.DataFrame(house_data)

# Model building
X_house = df_house[['House_Size']]
y_house = df_house['Price']

model_house = LinearRegression()
model_house.fit(X_house, y_house)

# Plotting
plt.scatter(X_house, y_house, color='green')
plt.plot(X_house, model_house.predict(X_house), color='orange')
plt.title("House Size vs Price")
plt.xlabel("Size (sq ft)")
plt.ylabel("Price ($1000)")
plt.show()
```
